- 크롤링 방식은 Request를 어떤 방식으로 하는가에 따라 다름
    - Web Browser를 이용하는 방식 : Selenium
    - Python 내에서 요청을 던지는 방식 : Requests, Urllibs
  
- 우리가 Request를 던지면 Response로 그 결과가 도착함. 
    - HTTP 규약을 기반으로 통신함.
    
- HTTP란?
    - HyperText Transfer Protocol > 웹 통신 과정인 Request, Response에 대한 규약
    - HTML, XML 등을 전송하는데 사용됨
    - 한번에 하나의 데이터만을 받을 수 있다는 특징이 있음. (원하는 데이터가 여러개라면 각각 요청해야함)




- Request 의 구성요소
    - Request는 포괄적으로 요청하는 것이 아니라, 원하는 바를 구체적으로 요청해야함. by URL
        <img src="Pic2.jpg">
    - GET : 서버에 파일이나 인덱스를 요청하는 과정
    - POST : 서버의 상태를 바꾸기 위한 요청과정
    - PUT, DELETE 등등
    - 크롤링에서 필요한 과정은 "Get"

- Response 
    - DOM, SAX 구조
    - DOM : HTML, XML document의 골격으로 넘어 옴. node들의 계층구조로 되어있음
        - tag + attribute 로 구성
        (https://developer.mozilla.org/ko/docs/Web/HTML/Element)
        
<img src="Pic3.jpg" width="700">

    - SAX : document를 하나의 line으로 받아옴. XML의 구조에서 이벤트 발생을 기준으로 처리하기 때문에 핸들러가 필요함 (저도 뭔말인지 모르겠어요)


    
- 파이썬 라이브러리 
    - Selenium : 웹브라우저를 사용, Ghost Driver...(하고 싶음)
    - Requests : 파이썬내에서 Request를 던짐. (Requests-HTML 이라는 추가 모듈로 Parsing도 함)
    - Urllib(3) : Request를 던질 수도 있고, Response로 온 데이터를 Parsing 하는 부분도 포함되어있음.

```python
import urllib3
from urllib.request import urlopen
```

```python
## Urllib3 에서 Requests를 사용하려면 PoolManager가 필요
urllib3.disable_warnings()
http = urllib3.PoolManager()
r = http.request(method = "GET", url = "https://news.naver.com/main/list.nhn?mode=LPOD&mid=sec&sid1=001&sid2=140&oid=001&isYeonhapFlash=Y/")
print(r.status)
r.data.decode('CP949')[15000:16000]
```

```python
html = urlopen("https://news.naver.com/main/list.nhn?mode=LPOD&mid=sec&sid1=001&sid2=140&oid=001&isYeonhapFlash=Y/")
html.read().decode("CP949")[15000:16000]
```

~~검색해서 타고 들어가는 과정을 셀레늄 처럼 따라할줄을 모르겠습니다..~~

그래서 간단한 네이버 평점 크롤러를 작성했습니다.

```python
pages = 10
for page in range(pages) :
    url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?type=after&order=newest&code=136900&page={}".format(page)
    res = requests.get(url)
    soup = BeautifulSoup(res.text,"html.parser")

    comments = soup.select("body > div > div > div.score_result > ul > li")
    for comment in comments:
        score = comment.select_one("div.star_score").text.strip()
        content = comment.select_one("div.score_reple > p").text.strip()
        author = comment.select_one("dl > dt > em").text.strip()
        date = comment.select_one("dl > dt > em:nth-of-type(2)").text.strip()
        up = comment.select_one("div.btn_area > strong:nth-of-type(1) > span").text.strip()
        down = comment.select_one("div.btn_area > strong:nth-of-type(2) > span").text.strip()
        print("====================================================================")
        print("댓글 : ",content)
        print("평점 : ", score)
        print("작성자 : ",author)
        print("작성일 : ",date)
        print("공감 : {} \n비공감 : {}".format(up, down))
        print("====================================================================")

    next = soup.find('a', {'class':"pg_next"})

```
